```python
import os
import getpass
from langchain.chat_models import init_chat_model

if not os.environ.get("GROQ_API_KEY"):
    os.environ["GROQ_API_KEY"] = getpass.getpass("Enter your Groq API key in environment vars ")

os.environ["LANGSMITH_TRACING"] = "true"
```

```python
from langchain.agents import create_agent
from langgraph.checkpoint.memory import InMemorySaver

agent = create_agent(
    model="groq:qwen/qwen3-32b",
    checkpointer=InMemorySaver()
)
```

```python
### Run with Thread ID
thread_01_config={
    "configurable":{
        "thread_id":"thread-001"
    } 
}
```

```python
from langchain.messages import SystemMessage, HumanMessage,AIMessage

response = agent.invoke({"messages":[HumanMessage(content="Hi, My name is Brijesh Dhaker !!!")]}, config=thread_01_config)
response
```

```python
response = agent.invoke({"messages":[HumanMessage(content="What is my name ?")]}, config=thread_01_config)
response
```
